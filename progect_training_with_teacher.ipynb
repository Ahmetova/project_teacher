## Исследование задачи

### Подготовка данных для выборок

Для начала необходимо создать две таблицы данных: таблица с признаками - `features`, таблица с целевым признаком - `target`.

Но перед этим подробнее разберемся с признаками.

**1.** Очевидно, что столбцы `row_number`, `customer_id` и `surname` не несут никакого информативного характера для определения, останется ли клиент в банке или уйдет. Скорее информация в них только еще больше запутает модель. Поэтому в таблицу с признаками данные столбцы не войдут.

**2.** В данных имеются как количественные признаки, так и категориальные. Следовательно, необходимо преобразовать категориальные признаки в численные. Используем технику прямого кодирования - OHE, с обходом дамми-ловушек.

```python
data_ohe = pd.get_dummies(data, drop_first=True, columns=['gender', 'geography'])
data_ohe.head()
```

**3.** Создадим две таблицы данных: таблица с признаками - `features`, таблица с целевым признаком - `target`.

```python
features = data_ohe.drop(['exited', 'row_number', 'customer_id', 'surname'], axis=1)
target = data['exited']

display('features:', features.head())
display('target:', target.head())

pd.options.mode.chained_assignment = None
```

Разобьем наши данные на три выборки: обучающую, валидационную и тестовую, в соотношении 3:1:1.

Для этого сначала разобъем данные на обучающую и тестовую выборки.

```python
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.4, random_state=12345)
```

Теперь разобъем тестовые данные на тестовую выборку и валидационную (делим пополам).

```python
features_test, features_valid, target_test, target_valid = train_test_split(
    features_test, target_test, test_size=0.5, random_state=12345)

print('Тренировочная:', features_train.shape)
print('Валидационная:', features_valid.shape)
print('Тестовая:', features_test.shape)
```

Таким образом, разбили имеющиеся данные на три выборки в соотношении 3:1:1.

**4.** Для того, что избежать ситуации, когда при обучении модели один какой-либо признак из-за наличия больших значений был засчитан как более значимый, в численных признаках в столбцах `credit_score`, `age`, `tenure`, `balance`, `num_of_products` и `estimated_salary` необходимо провести масштабирование. Для этого воспользуемся методом стандаризации данных.

```python
scaler = StandardScaler()

numeric = ['credit_score', 'age', 'tenure', 'balance', 'num_of_products', 'estimated_salary']

scaler.fit(features_train[numeric]) 

features_train[numeric] = scaler.transform(features_train[numeric])
features_valid[numeric] = scaler.transform(features_valid[numeric])
features_test[numeric] = scaler.transform(features_test[numeric])

display('features_train:', features_train.head())
display('features_valid:', features_valid.head())
display('features_test:', features_test.head())
```

Посмотрим, как будут вести себя модели, если не сбалансировать классы для целевого признака.

#### Модель решающего дерева

Будем подбирать гиперпараметр `max_depth` в цикле.

```python
best_model_Tree = None
best_f1_Tree = 0
best_depth = 0

start_time = time.perf_counter()

for depth in range(1, 100):
    model = DecisionTreeClassifier(random_state=12345, max_depth=depth) # модель
    model.fit(features_train, target_train) # обучение модели
    predictions_valid_Tree = model.predict(features_valid) # предсказания 
    f1 = f1_score(target_valid, predictions_valid_Tree)
    if best_f1_Tree < f1:
        best_model_Tree = model
        best_f1_Tree = f1
        best_depth = depth

end_time = time.perf_counter()
elapsed_time_Tree = end_time - start_time

print('Гиперпараметры:', end='\n')
print("best_depth =", best_depth)
print()
print("best_f1_Tree =", best_f1_Tree) 
print('Время работы программы =', elapsed_time_Tree, 'сек')
```

Лучшее значение `f1_score` получаем при глубине дерева равной 6.

#### Модель случайного леса

Подберем гиперпараметры `max_depth` и `n_estimators` в цикле.

```python
best_model_Forest = None
best_f1_Forest = 0
best_est = 0
best_depth = 0

start_time = time.perf_counter()

for est in range(10, 50):
    for depth in range (1, 25):
        model = RandomForestClassifier(random_state=12345, n_estimators=est, max_depth=depth) # модель
        model.fit(features_train, target_train) # обучение модель 
        predictions_valid_Forest = model.predict(features_valid) # предсказания модели 
        f1 = f1_score(target_valid, predictions_valid_Forest)
        if best_f1_Forest < f1:
            best_model_Forest = model
            best_f1_Forest = f1
            best_est = est
            best_depth = depth
            
end_time = time.perf_counter()
elapsed_time_Forest = end_time - start_time

print('Гиперпараметры:', end='\n')
print("best_depth =", best_depth, end='\n')
print("best_n_estimators =", best_est, end='\n')
print()
print("best_f1_Forest =", best_f1_Forest) 
print('Время работы программы =', elapsed_time_Forest, 'сек')
```

Лучшее значение `f1_score` получаем при количестве деревьев равное 31 и глубине каждого дерева равного 20.

```python
fpr, tpr, thresholds = roc_curve(target_test, probabilities_one_test)

plt.figure()
plt.plot(fpr, tpr)  
plt.plot([0, 1], [0, 1], linestyle='--')
plt.ylim([0.0, 1.0])
plt.xlim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая')

plt.show()
```
